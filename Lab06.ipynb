```python
import pandas as pd

# 1. Create a sample sales dataset with duplicate entries
data = {
    'OrderID': [1, 2, 3, 4, 5, 1, 6, 7, 8, 4],
    'CustomerID': ['A', 'B', 'C', 'D', 'E', 'A', 'F', 'G', 'H', 'D'],
    'Product': ['X', 'Y', 'Z', 'X', 'Y', 'X', 'Z', 'Y', 'X', 'X'],
    'Price': [100, 150, 200, 100, 150, 100, 200, 150, 100, 100]
}

sales_df = pd.DataFrame(data)

print("Original DataFrame:")
display(sales_df)

# 2. Identify duplicate records
duplicate_rows = sales_df[sales_df.duplicated()]
print("\nDuplicate rows:")
display(duplicate_rows)

print("\nNumber of duplicate rows before removal:", duplicate_rows.shape[0])

# 3. Remove duplicate records
sales_df_cleaned = sales_df.drop_duplicates()

# 4. Verify duplicate removal
print("\nDataFrame after removing duplicates:")
display(sales_df_cleaned)

print("\nNumber of rows after removal:", sales_df_cleaned.shape[0])

# Verify that no duplicates remain
print("\nNumber of duplicate rows after removal:", sales_df_cleaned[sales_df_cleaned.duplicated()].shape[0])
```

```python
import pandas as pd

# 1. Create a sample employee dataset with incorrect data types
data = {
    'EmployeeID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': ['50000', '60000', '75000', '55000', '62000'], # Salary as string
    'Joining_Date': ['2022-01-15', '2021-11-20', '2023-03-10', '2022-07-01', '2023-01-30'], # Joining_Date as string
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT']
}

employee_df = pd.DataFrame(data)

print("Original DataFrame:")
display(employee_df)

# 2. Identify initial data types
print("\nInitial data types:")
display(employee_df.dtypes)

# 3. Convert Salary to numeric
employee_df['Salary'] = pd.to_numeric(employee_df['Salary'])

# 4. Convert Joining_Date to datetime
employee_df['Joining_Date'] = pd.to_datetime(employee_df['Joining_Date'])

# 5. Verify data type conversion
print("\nData types after conversion:")
display(employee_df.dtypes)

print("\nDataFrame after type conversion:")
display(employee_df)
```
