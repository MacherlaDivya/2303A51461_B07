# Task
Generate Python code to create two datasets (Teachers and Courses), merge them using inner and left joins on `teacher_id`, compute the average courses per teacher, and group by department to calculate the total workload. Discuss the findings on teacher workload and department-wise distribution.

## Create datasets

### Subtask:
Create two pandas DataFrames, one for Teachers and one for Courses, with appropriate columns and sample data.


**Reasoning**:
Create the teachers and courses data using dictionaries and then convert them to pandas DataFrames as per the instructions.



```python
teachers_data = {
    'teacher_id': [1, 2, 3, 4, 5],
    'teacher_name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'department': ['Math', 'Physics', 'Math', 'Chemistry', 'Physics']
}

courses_data = {
    'course_id': [101, 102, 103, 104, 105, 106, 107],
    'course_name': ['Calculus', 'Mechanics', 'Linear Algebra', 'Organic Chemistry', 'Thermodynamics', 'Electromagnetism', 'Differential Equations'],
    'teacher_id': [1, 2, 1, 4, 2, 2, 3],
    'credits': [4, 3, 4, 4, 3, 3, 4]
}

teachers_df = pd.DataFrame(teachers_data)
courses_df = pd.DataFrame(courses_data)

print("Teachers DataFrame:")
display(teachers_df)

print("\nCourses DataFrame:")
display(courses_df)
```

## Perform inner merge

### Subtask:
Merge the Teachers and Courses DataFrames using an inner join on `teacher_id`.


**Reasoning**:
Merge the teachers_df and courses_df DataFrames using an inner join on the teacher_id column and display the result.



```python
inner_merged_df = pd.merge(teachers_df, courses_df, on='teacher_id', how='inner')
display(inner_merged_df)
```

## Perform left merge

### Subtask:
Merge the Teachers and Courses DataFrames using a left join on `teacher_id`, keeping all teachers even if they have no courses.


**Reasoning**:
Merge the teachers_df and courses_df DataFrames using a left join on 'teacher_id', keeping all teachers.



```python
left_merged_df = pd.merge(teachers_df, courses_df, on='teacher_id', how='left')
display(left_merged_df)
```

## Compute average courses per teacher

### Subtask:
Calculate the average number of courses taught per teacher.


**Reasoning**:
Group the inner_merged_df by teacher_id to count the courses per teacher and then calculate the average number of courses.



```python
courses_per_teacher = inner_merged_df.groupby('teacher_id').size()
average_courses_per_teacher = courses_per_teacher.mean()
print(f"Average number of courses per teacher: {average_courses_per_teacher:.2f}")
```

## Group by department and calculate total workload

### Subtask:
Group the merged DataFrame by department and calculate the total number of courses taught in each department.


**Reasoning**:
Group the merged dataframe by department and calculate the number of courses in each department.



```python
department_workload = inner_merged_df.groupby('department').size()
display(department_workload)
```

**Reasoning**:
Based on the calculated average courses per teacher and the department workload, write a summary of the findings, discussing the average workload and the distribution across departments.



```python

print(f"The average number of courses taught per teacher is: {average_courses_per_teacher:.2f}")

display(department_workload)

print("\nSummary of Findings:")
print(f"On average, teachers teach {average_courses_per_teacher:.2f} courses.")

```
