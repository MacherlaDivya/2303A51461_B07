```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
try:
    df_sales = pd.read_csv('sales_data.csv')
except FileNotFoundError:
    print("Error: 'sales_data.csv' not found. Please upload the file or provide the correct path.")
    # Create a dummy dataframe for demonstration if the file is not found
    data_sales = {'Quarter': ['Q1', 'Q2', 'Q3', 'Q4', 'Q1', 'Q2', 'Q3', 'Q4'],
                  'Sales': [150000, 200000, 250000, 300000, 160000, 210000, 260000, 310000]}
    df_sales = pd.DataFrame(data_sales)
    print("Using dummy sales data for demonstration.")

# Plot the quarterly sales
plt.figure(figsize=(10, 6))
plt.plot(df_sales['Quarter'], df_sales['Sales'], marker='o', linestyle='-')

# Highlight peak sales
peak_sales_quarter = df_sales.loc[df_sales['Sales'].idxmax()]
plt.plot(peak_sales_quarter['Quarter'], peak_sales_quarter['Sales'], marker='*', markersize=10, color='red', label=f"Peak Sales: {peak_sales_quarter['Sales']:,} ({peak_sales_quarter['Quarter']})")


# Add grid lines and labels
plt.grid(True)
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Quarterly Sales Performance')
plt.legend()

plt.tight_layout() # Adjust layout to prevent labels overlapping

# Display the plot
plt.show()

# Discussion of seasonal sales trends (Based on dummy data or assuming typical retail trends)
print("\nDiscussion of Seasonal Sales Trends:")
print("Based on the plot, we can observe the quarterly sales trend.")
print(f"The peak sales occurred in {peak_sales_quarter['Quarter']} with sales of {peak_sales_quarter['Sales']:,}.")
print("Generally, retail sales often show seasonality, with higher sales during holiday seasons (e.g., Q4).")
print("Analyzing the trend over multiple years would provide a clearer picture of consistent seasonal patterns.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
try:
    df = pd.read_csv('daily_temperature.csv')
except FileNotFoundError:
    print("Error: 'daily_temperature.csv' not found. Please upload the file or provide the correct path.")
    # Create a dummy dataframe for demonstration if the file is not found
    data = {'Day': range(1, 31),
            'Temperature': [20 + i*0.5 + (i%7)*2 - (i%5)*1.5 for i in range(30)]}
    df = pd.DataFrame(data)
    print("Using dummy data for demonstration.")


# Plot the temperature variation
plt.figure(figsize=(10, 6))
plt.plot(df['Day'], df['Temperature'], marker='o', linestyle='-')

# Add grid lines and labels
plt.grid(True)
plt.xlabel('Day')
plt.ylabel('Temperature (°C)')
plt.title('Daily Temperature Variation Over a Month')
plt.xticks(df['Day']) # Ensure all days are shown on the x-axis
plt.tight_layout() # Adjust layout to prevent labels overlapping

# Display the plot
plt.show()

# Interpret hottest and coldest days
hottest_day = df.loc[df['Temperature'].idxmax()]
coldest_day = df.loc[df['Temperature'].idxmin()]

print(f"\nHottest Day: Day {hottest_day['Day']} with temperature {hottest_day['Temperature']}°C")
print(f"Coldest Day: Day {coldest_day['Day']} with temperature {coldest_day['Temperature']}°C")
```
